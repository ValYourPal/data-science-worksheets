```r
#Question 2.2
fast_food <- read_csv("data/fast_food.csv")

fast_food
test_2.2()
```

```r
#Question 2.3
top_restaurants <- fast_food %>%
    filter(st %in% c("CA", "WA", "OR")) %>% 
    group_by(name) %>%
    summarize(n = n()) %>%
    arrange(desc(n)) %>% 
    head(n = 9)

top_restaurants
test_2.3()
```

```r
#Question 2.4
count_bar_chart <- ggplot(data = top_restaurants, aes(x = name, y = n)) + 
    geom_bar(stat = "identity") +
    xlab("Name of Restaurant") +
    ylab("Number of Restaurants") +
    theme(text = element_text(size = 20))


count_bar_chart
test_2.4()
```

```r
#Question 2.5
#PART A
count_bar_chart_A <- count_bar_chart + 
    theme(axis.text.x = element_text(angle = 80, hjust = 1))

count_bar_chart_A

#PART B
count_bar_chart_B <- count_bar_chart + 
    coord_flip()

count_bar_chart_B
test_2.5_A()
test_2.5_B()
```

```r
#Question 2.7
state_counts <- fast_food %>%
    semi_join(top_restaurants) %>% # semi_join gives the intersection of two data frames
    filter(st %in% c("CA", "WA", "OR")) %>% 
    group_by(st) %>%
    summarize(n = n()) 

state_counts
test_2.7()
```

```r
#Question 2.8
options(repr.plot.width = 15, repr.plot.height = 10) 

state_counts_plot <- ggplot(data = state_counts, aes(x = st, y = n)) + 
    geom_bar(stat = "identity") +
    xlab("State") +
    ylab("Number of Restaurants") +
    theme(text = element_text(size = 20))
  
state_counts_plot
test_2.8()
```

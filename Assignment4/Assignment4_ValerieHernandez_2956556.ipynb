```r
#Question 3.2
madrid <- read_tsv("data/madrid_pollution.csv")
madrid 
test_3.2()
```

```r
#Question 3.3
options(repr.plot.width = 13, repr.plot.height = 7)

EBE_pollution <- madrid %>% 
    ggplot(aes(x = date, y = EBE)) + 
    geom_point() + 
   xlab("Date") + 
    ylab("Ethylbenzene (Œºg/m¬≥)") +
    theme(text = element_text(size=20))

EBE_pollution

# Are levels increasing or decreasing?
# My answer: They are decreasing
test_3.3()
```

```r
?group_by
```

```r
#Question 3.4
madrid_pollution <-  madrid %>%
     group_by(year, mnth) %>%
     summarize(max_ebe = max(EBE, na.rm = TRUE))

madrid_pollution
test_3.4()
```

```r
#Question 3.5
madrid_plot <- madrid_pollution %>%
    ggplot(aes(x = mnth, y = max_ebe)) + 
    geom_point() +
    xlab("Month") + 
    ylab("Maximum Ethylbenzene (Œºg/m¬≥)") +
    facet_grid(~ year) +
    theme(axis.text.x = element_text(angle = 90, hjust = 1)) +
    theme(text = element_text(size=20))

madrid_plot
test_3.5()
```

```r
#Question 3.6
pollution_2001 <- madrid %>%
     filter(year == 2001) %>%
     select(-date, -year, -mnth) %>%
     map_df(max, na.rm  = TRUE)

pollution_2001
test_3.6()
```

```r
#Question 3.7
pollution_2006 <- madrid %>%
     filter(year == 2006) %>%
     select(-date, -year, -mnth) %>%
     map_df(max, na.rm  = TRUE)

pollution_2006
test_3.7()
```

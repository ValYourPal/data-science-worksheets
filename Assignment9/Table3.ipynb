```r
#Question 1.8
beer_ks <- tibble(k = 1:10)

beer_ks
test_1.8()
```

```r
#Question 1.9
set.seed(1234) # DO NOT REMOVE
beer_clustering <- beer_ks %>%
     mutate(models = map(k, ~kmeans(scaled_beer, .x)))
   
print(beer_clustering)
test_1.9()
```

```r
#Question 2.0
beer_model_stats <- beer_clustering %>% 
  mutate(model_statistics = map(models, ~glance(.x)))

print(beer_model_stats)
test_2.0()
```

```r
beer_model_stats %>% 
    slice(1) %>% 
    pull(model_statistics)
```

```r
#Question 2.1
beer_clustering_unnested <- beer_model_stats %>% unnest(model_statistics)

print(beer_clustering_unnested)
test_2.1()
```

```r
#Question 2.2

options(repr.plot.width = 8, repr.plot.height = 7)

choose_beer_k <- beer_clustering_unnested %>%
  ggplot(aes(x = k, y = tot.withinss)) +
  geom_point() +
  geom_line() +
  labs(x = "K", y = "Total Within-Cluster Sum of Squares") +
  theme_minimal()

choose_beer_k
test_2.2()
```
